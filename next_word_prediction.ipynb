#  Make Pairs

```python
def get_pairs(words):
    pairs = []
    for i in range(len(words)-1):
        pairs.append(" ".join(words[i : i+2]))
    return pairs
get_pairs(words)
```

#  Finding Occurence Probability

```python
a = np.array(get_pairs(words[:300000]))
```

```python
pair , count = np.unique(a , return_counts = True)
```

```python
print(len(pair))
print(len(count))
```

```python
prob_dist = []
for i in range(len(pair)):
    prob_dist.append([pair[i] , count[i] , pair[i].split(" ")[-1]])
print(len(prob_dist))
```

#  predict word

```python
data = pd.DataFrame(prob_dist , columns = ["inp" , "freq" , "out"])
data = data[data["freq"] >= 5]
def predict_word(word):

    predicted_words = []
    for i in data.values:
        if (i[0].split(" ")[0] == word):
            predicted_words.append([i[0] , i[1] , i[2]])
    predicted_words = pd.DataFrame(predicted_words , columns = ["imput" , "freq" , "output"])
    return (list(predicted_words.sort_values(by = "freq" , ascending = False).head()["output"].values))
word = input("enter your word here")
print("predicted next words are :",predict_word(word))


# def predict_word(word):
#     predicted_words = []
#     for i in data.values:
#         print(i.split(","))
        
# predict_word()   
```
